```python
# Creating a final data table for hypothesis testing
_results = new_results.merge(users, on='user_id', how='left')
del _results['country']
del _results['lat']
del _results['long']

# creating city catagories New York, Chicago, Houston, Others and NA (Not Available)
def filter_city(r):
    if isinstance(r['city'], float):
        city = 'NA'
    elif r['city'] == 'New York' or r['city'] == 'Chicago' or r['city'] == 'Houston':
        city = r['city']
    else:
        city = 'Other'
    try:
        datetime = pd.Timestamp(r['timestamp'])
    except:
        datetime = None
    return pd.Series({'datetime':datetime, 'source':r['source'], 'device':r['device'], 'operative_system':r['operative_system'], 'test':r['test'], 'converted':r['converted'], 'city':city})
final_result = _results.apply(filter_city, axis=1).dropna()
```

# Hypothesis Testing

H_o: There is a no statisticaly significant change in conversion rate with increase in price

Using chi^2 test in the following:

```python
conversion_table = pd.concat([final_result[final_result.test==0]['converted'].value_counts().rename('control'), final_result[final_result.test==1]['converted'].value_counts().rename('test')],axis=1).transpose()
conversion_table.columns = ['not_converted', 'converted']
conversion_table
```

```python
from scipy.stats import chi2_contingency

chi2_contingency(conversion_table)[:2]
```

As p value << 0.05, we can neglect the null hypothesis, and thus, there is a statistically significant change in the conversion rate after increase in price. The conversion rate in control group is around 2.0% while that in the test group is 1.55%. Hence there is an overall decrease in the conversion rate with increase in the price to $59.

# Optimum Sample Size

```python
# function to get p-value for dataframe d
def get_p(df):
    conversion_table = pd.concat([df[df.test==0]['converted'].value_counts().rename('control'), df[df.test==1]['converted'].value_counts().rename('test')],axis=1).transpose()
    conversion_table.columns = ['not_converted', 'converted']
    return chi2_contingency(conversion_table)[:2][1]

def get_date_p_data(df):
    l=[]
    count = df.size
    for i in range(count-1000, count):
        l.append((df.head(i).tail(1)['date'].values[0], df.head(i).tail(1)['time'].values[0], get_p(df.head(i))))
    return l

final_result_sorted = final_result.sort_values(['datetime'])
data_x = []
data_y = []
for i in range(1, final_result_sorted['datetime'].size, 100):
    data_x.append(final_result_sorted[:-i].tail(1)['datetime'].values[0])
    data_y.append(get_p(final_result_sorted[:-i]))
```

```python
plt.plot(data_x[2100:], data_y[2100:])
plt.plot(data_x[2100:], np.array([0.05 for i in data_y[2100:]]), 'r--')
plt.title('p value - time plot')
plt.xlabel('DateTime')
plt.ylabel('p value')
plt.show()
```

The study could have been ended after first week as the p value decreses from 0.05 after that.

# Conclusion

There is a statisitcally significant decrease in the conversion rate with increase in product price to \$59 from \$39. Now checking revenue:

```python
def get_revenue(r):
    if r.name == 'control':
        rev = 39.0 * r['converted']
        rev_p_v = 39.0 * r['converted'] / (r['converted'] + r['not_converted'])
    if r.name == 'test':
        rev = 59.0 * r['converted']
        rev_p_v = 59.0 * r['converted'] / (r['converted'] + r['not_converted'])
    return pd.Series({'not_converted':r['not_converted'], 'converted':r['converted'], 'revenue':rev, 'revenue_per_visit':rev_p_v})

conversion_table.apply(get_revenue, axis=1)
```

It can be concluded that the company will have more revenue per number of user visits with increase in price!
