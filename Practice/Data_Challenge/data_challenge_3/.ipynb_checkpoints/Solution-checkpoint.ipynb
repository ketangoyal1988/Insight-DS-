```python
model_data=data[data.age<100]
```

# Modeling

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold, train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import preprocessing as pre

X, y = model_data[model_data.columns[:-1]], model_data['converted']

# Encoding catagorical data for model training
le_country = pre.LabelEncoder()
le_source = pre.LabelEncoder()

X['country'] = le_country.fit_transform(X['country'])
X['source'] = le_source.fit_transform(X['source'])

X.describe()
```

```python
X.corr()
```

Features are independent of each other.

```python
# Splitting data into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

print('Random Forest....')
gs_rf = GridSearchCV(
    estimator=RandomForestClassifier(),
    param_grid={'n_estimators': [40,100], 
                'class_weight': [None, 'balanced', 'balanced_subsample'], 
                'bootstrap': [True, False],
                'max_depth': [3, None],
                'criterion': ['gini', 'entropy']},
    cv=StratifiedKFold(n_splits=9),
    n_jobs=4,
    )
model_rf = gs_rf.fit(X_train, y_train)
print('Done')
```

```python
model_rf.best_estimator_
```

```python
from sklearn.metrics import confusion_matrix, precision_recall_curve

y_prob_rf = model_rf.predict_proba(X_test)
precisions, recalls, thresholds = precision_recall_curve(y_test, y_prob_rf[:,1])

plt.step(recalls, precisions, color='b', alpha=0.2, where='post')
plt.fill_between(recalls, precisions, step='post', alpha=0.2, color='b')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('Precision-Recall Curve')
plt.show()
```

```python
plt.plot(thresholds, precisions[:-1], label='Precision')
plt.plot(thresholds, recalls[:-1], label='Recall')
plt.xlabel('Threshold')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
from sklearn.metrics import roc_auc_score, roc_curve, auc

fpr, tpr, thresholds = roc_curve(y_test, y_prob_rf[:,1])

plt.plot(fpr, tpr, lw=2, label='ROC (AUC = %0.2f)' % (auc(fpr, tpr)))
plt.plot([0, 1], [0, 1], linestyle='--',color='r')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
plt.plot(thresholds, fpr, label='FPR')
plt.plot(thresholds, tpr, label='TPR')
plt.xlabel('Threshold')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
importances = model_rf.best_estimator_.feature_importances_
std = np.std([tree.feature_importances_ for tree in model_rf.best_estimator_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(X.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, f, importances[f]))

# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(range(X.shape[1]), importances[range(importances.size)],
       color="r", yerr=std[range(importances.size)], align="center")
plt.xticks(range(X.shape[1]), X.columns)
plt.xlim([-1, X.shape[1]])
plt.show()
```

As predicted by the model, the total pages visited is the most predictive feature followed by someone being a new user, thier country and age.

# Conclusion

The most predicitive feature for conversion in the model is the total pages visited by a user. Furthermore, a new user is more likely to get converted compared to an exsisting user. The marketing channel does not seam to play a role in conversion as one would be likely to buy an item irrespective of how they reached the site. Thus to increase revenue, the product retailers should try to attract as many new and young users as possible.
